```
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
xray01,xray02,xray03,xray04= cv2.imread('xray01.png',0),cv2.imread('xray02.png',0),cv2.imread('xray03.png',0),cv2.imread('xray04.png',0)

  
```

```
def gamma_corr(img,gamma):
  tmp=img/255
  corrected_image=np.rint(255*pow(tmp,gamma))
  return corrected_image
gamm_corr01=gamma_corr(xray01,2.5)
gamm_corr02=gamma_corr(xray02,2.7)
gamm_corr03=gamma_corr(xray03,3)
gamm_corr04=gamma_corr(xray04,2.5)
```

```
def hstgrm(img):
  intnsty_bins=np.zeros((256,1))
  im=np.reshape(img,(1,np.size(img)))
  for x in im[0,:]:
    intnsty_bins[x]+=1
  return (intnsty_bins/np.size(img)).T
  
hist01=hstgrm(xray01)
hist02=hstgrm(xray02)
hist03=hstgrm(xray03)
hist04=hstgrm(xray04)
  
```

```
def hstgrm_equl(hist_img,img):
  equalized_hist=np.zeros((1,256))
  for x in range(np.size(hist_img)):
    equalized_hist[0,x]=sum(hist_img[0,0:x+1])
  new_img = np.zeros((1024,1024))
  for i in range(1024):
    for j in range(1024):
      a = img[i,j]
      b = equalized_hist[0,a]
      b = b*255
      b = int(b+0.5)
      new_img[i,j] = b
  return new_img

hstgrm_equl01=hstgrm_equl(hist01,xray01)
hstgrm_equl02=hstgrm_equl(hist02,xray02)
hstgrm_equl03=hstgrm_equl(hist03,xray03)
hstgrm_equl04=hstgrm_equl(hist04,xray04)
```

```
def AHE(img,win_size):
  s=int((win_size-1)/2)
  pad=np.pad(img,((s,s),(s,s)),'constant')
  new_img=np.zeros((1024,1024))
  for i in range(0,1024):
    for j in range(0,1024):
      hst=np.zeros((1,256))/win_size**2
      u=np.reshape(pad[i:win_size,j:win_size],(1,np.size(pad[i:win_size,j:win_size])))
      for x in u[0]:
        hst[0,u]+=1
      equalized_hist=np.zeros((1,256))
      for y in range(np.size(hst)):
          equalized_hist[0,y]=sum(hst[0,0:y+1])
          a = img[i,j]
          b = equalized_hist[0,a]
          b = b*255
          b = int(b+0.5)
          new_img[i,j] = b
  return new_img    
        
      
```
